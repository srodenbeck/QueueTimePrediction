```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import psycopg2
```

```python
path_root = "datafiles/"
f_type = ".csv"

months = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]
years = ["2022_data", "2023_data", "2024_data"]

monthly_data = {}

for month in months:
    end = False
    for year in years:
        if year == "2024_data" and month not in months[0:3]:
            break
        path = path_root + month + year + f_type
        month_year_data = month + year
        monthly_data[month_year_data] = pd.read_csv(path, delimiter="|")
      
        print(monthly_data[month_year_data].head())
```

```python
for month in months:
    end = False
    for year in years:
        if year == "2024_data" and month not in months[0:3]:
            break
        path = path_root + month + year + f_type
        month_year_data = month + year
        monthly_data[month_year_data] = pd.read_csv(path, delimiter="|")
        
                # Converting strings into pandas datetime format
        monthly_data[month_year_data]['Start'].replace('Unknown', pd.NaT, inplace=True)
        monthly_data[month_year_data]['Submit'].replace('Unknown', pd.NaT, inplace=True)
        monthly_data[month_year_data]['End'].replace('Unknown', pd.NaT, inplace=True)
        # monthly_data[month_year_data]['ElapsedRaw'].replace('Unknown', pd.NaT, inplace=True)
        # Adding a 'Delay' column which could be used to 
        monthly_data[month_year_data]['Start'] = pd.to_datetime(monthly_data[month_year_data]['Start'])
        monthly_data[month_year_data]['Submit'] = pd.to_datetime(monthly_data[month_year_data]['Submit'])
        # monthly_data[month_year_data]['ElapsedRaw'] = pd.to_timedelta(monthly_data[month_year_data]['ElapsedRaw'], unit='s')
        monthly_data[month_year_data]['End'] = pd.to_datetime(monthly_data[month_year_data]['End'])
        monthly_data[month_year_data]['Delay'] = monthly_data[month_year_data]['Start'] - monthly_data[month_year_data]['Submit']
        monthly_data[month_year_data]['DelayRaw'] = monthly_data[month_year_data]['Delay'].view('int64') // 10**9

```

```python
class DataChart():
    def __init__(self, month, min=0, max=0, mean=0, count=0):
        self.month = month
        self.min = min
        self.max = max
        self.mean = mean
        self.count = count
    def __str__(self):
        return self.month + " Min: " + str(self.min) + ", Max: " + str(self.max) + ", Mean: " + str(self.mean) + ", Count: " + str(self.count)
```

```python
column_name = "ElapsedRaw"

my_data = []

print(column_name)

for month in months:
    end = False
    for year in years:
        if year == "2024_data" and month not in months[0:3]:
            break
        
        all_values = []
        df = monthly_data[month + year]
        filtered_values = df[df["State"] == "COMPLETED"][column_name].astype(int)
        all_values.extend(filtered_values)
        
        mean_value = sum(all_values) / len(all_values)
        min_value = min(all_values)
        max_value = max(all_values)
        num_values = len(all_values)
        
        my_data.append(DataChart(month + year, min_value, max_value, mean_value, num_values))
```

```python
for i in range(len(my_data)):
    for j in range(len(my_data) - 1):
        year_1 = int(my_data[j].month[6])
        year_2 = int(my_data[j+1].month[6])
        month_1 = my_data[j].month[0:3]
        month_2 = my_data[j+1].month[0:3]
        
        if year_1 > year_2:
            temp = my_data[j]
            my_data[j] = my_data[j+1]
            my_data[j+1] = temp
        elif year_1 < year_2:
            continue
        else:
            if months.index(month_1) > months.index(month_2):
                temp = my_data[j]
                my_data[j] = my_data[j+1]
                my_data[j+1] = temp
```

```python

### JOB RUN TIME

min_values = [data.min for data in my_data]
max_values = [data.max for data in my_data]
mean_values = [data.mean for data in my_data]
count_values = [data.count for data in my_data]

x_labels = list(range(1, 28))

def plot_best_fit(ax, x, y):
    coeffs = np.polyfit(x, y, 1)
    poly = np.poly1d(coeffs)
    y_fit = poly(x)
    ax.plot(x, y_fit, color='red', linestyle='--', label='Best Fit Line')
    ax.legend()

fig, axs = plt.subplots(4, 1, figsize=(10, 15))

# Plot min values
axs[0].plot(x_labels, min_values, marker='o', linestyle='-')
axs[0].set_title('Min Values of Runtime by Month')
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Min (s)')
plot_best_fit(axs[0], x_labels, min_values)

# Plot max values
axs[1].plot(x_labels, max_values, marker='o', linestyle='-')
axs[1].set_title('Max Values of Runtime by Month')
axs[1].set_xlabel('Month')
axs[1].set_ylabel('Max (s)')
plot_best_fit(axs[1], x_labels, max_values)

# Plot mean values
axs[2].plot(x_labels, mean_values, marker='o', linestyle='-')
axs[2].set_title('Mean Values of Runtime by Month')
axs[2].set_xlabel('Month')
axs[2].set_ylabel('Mean (s)')
plot_best_fit(axs[2], x_labels, mean_values)

# Plot count values
axs[3].plot(x_labels, count_values, marker='o', linestyle='-')
axs[3].set_title('Count Values of Runtime by Month')
axs[3].set_xlabel('Month')
axs[3].set_ylabel('Count')
plot_best_fit(axs[3], x_labels, count_values)

# Adjust layout
plt.tight_layout()
plt.show()
```
